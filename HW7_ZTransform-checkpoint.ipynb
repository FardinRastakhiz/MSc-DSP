```python
import numpy as np
import scipy
from scipy import signal, integrate
import matplotlib.pyplot as plt
from matplotlib import style
import lcapy as lc
from lcapy.discretetime import n as nn
```

```python
style.use('ggplot')
style.use('dark_background')
```

```python
xt = lambda t: (1 / (1+np.exp(-6*t)))*np.exp(-t)
number_of_samples = 200
time_domain_bound = 20
transform_domain_bound = 25
times = np.linspace(-time_domain_bound, time_domain_bound, number_of_samples)
xn = xt(times)

xn_sympy = (1 / (1+lc.exp(-6*nn)))*lc.exp(-nn)
```

```python
compz = complex(0, 1)
w = np.linspace(-transform_domain_bound, transform_domain_bound, number_of_samples)
r = 1
sampling_rate_w = (transform_domain_bound - (-transform_domain_bound))/number_of_samples
HZ1 = np.zeros((number_of_samples), dtype=complex)
z = lambda r, k: r*np.exp(compz * w[k])
for k in range(number_of_samples):
    for n in range(number_of_samples):
        HZ1[k] += xn[n]*(z(r, k)**-n)
    HZ1[k] *= sampling_rate_w
```

```python
HZ2 = np.fft.fft(xn)
```

```python
HZ3  = xn_sympy.ZT()
```

```python
HZ3
```

```python
v1 = HZ3(z(1, np.arange(number_of_samples)))
```

```python
v2 = np.array([v.evalf().evaluate() for v in v1])
```

```python
plt.plot(w, np.real(HZ1), color='cyan', label='z transform')
plt.plot(w, np.real(HZ2), color='yellow', label='fast fourier transform')
plt.plot(w, np.real(v2), color='magenta', label='lcapy z transform')
plt.legend()
plt.show()
```

```python
def inverse_z_transform(hz):
    xn_new = np.zeros((number_of_samples), dtype=complex)
    sampling_rate_time = (time_domain_bound - (-time_domain_bound))/number_of_samples
    for n in range(number_of_samples):
        real_hz = lambda w_id: np.real(hz[w_id] * (z(r, w_id) ** n))
        imag_hz = lambda w_id: np.imag(hz[w_id] * (z(r, w_id) ** n))
        real_hz_values = real_hz(np.arange(number_of_samples))
        imag_hz_values = imag_hz(np.arange(number_of_samples))
        x_real = integrate.simpson(real_hz_values)
        x_imag = integrate.simpson(imag_hz_values)
        xn_new[n] = sampling_rate_time * (x_real + compz*x_imag) #/ (2*np.pi)
    return xn_new
```

```python
def lcapy_inverse_z_transform(hz):
    ihz = hz[nn]
    xn_new = np.array([ihz.evalf().evaluate() for v in v1])
    return xn_new
```

```python
xn1 = inverse_z_transform(HZ1) 
xn2 = np.fft.ifft(HZ2)
xn3 = inverse_z_transform(v2) 

xn1
```

```python
plt.ylim(-1, 3)
plt.subplot(3, 1, 1)
plt.plot(times, np.real(xn1), color='magenta', label='written discrete time fourier transform')
plt.plot(times, np.real(xn)+0.1, color='cyan', label='original function (+0.1)')
plt.legend()
plt.subplot(3, 1, 2)
plt.plot(times, np.real(xn2), color='yellow', label='fast fourier transform')
plt.plot(times, np.real(xn)+0.1, color='cyan', label='original function (+0.1)')
plt.legend()
plt.subplot(3, 1, 3)
plt.plot(times, np.real(xn3), color='yellow', label='lcapy z transform')
plt.plot(times, np.real(xn)+0.1, color='cyan', label='original function (+0.1)')
plt.legend()
plt.show()
```
