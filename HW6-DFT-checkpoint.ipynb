```python
compz = complex(0, 1)

kw = np.arange(number_of_samples)
w = (2 * np.pi * kw)/number_of_samples
HZ1 = np.zeros((number_of_samples), dtype=complex)
for k in range(number_of_samples):
    for n in range(number_of_samples):
        HZ1[k] += xn[n]*np.exp(-compz * w[k] * n)
```

```python
HZ2 = np.fft.fft(xn)
```

```python
plt.plot(w, np.real(HZ1), color='cyan', label='discrete time fourier transform')
plt.plot(w, np.real(HZ2)+10, color='yellow', label='fast fourier transform (+10)')
plt.legend()
plt.show()
```

```python
def inverse_discrete_fourier_transform(hz):
    xn_new = np.zeros((number_of_samples), dtype=complex)
    for n in range(number_of_samples):
        real_hz = lambda w_id: np.real(hz[w_id] * np.exp(compz * w[w_id] * n))
        imag_hz = lambda w_id: np.imag(hz[w_id] * np.exp(compz * w[w_id] * n))
        real_hz_values = real_hz(np.arange(number_of_samples))
        imag_hz_values = imag_hz(np.arange(number_of_samples))
        x_real = integrate.simpson(real_hz_values)
        x_imag = integrate.simpson(imag_hz_values)
        xn_new[n] = (x_real + compz*x_imag) / number_of_samples # / (2*np.pi)
    return xn_new
```

```python
xn1 = inverse_discrete_fourier_transform(HZ1) 
xn2 = np.fft.ifft(HZ2)
```
