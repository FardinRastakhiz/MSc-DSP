```python
import numpy as np
import scipy
from scipy import signal, integrate
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
style.use('ggplot')
style.use('dark_background')
```

```python
xt = lambda t: (1 / (1+np.exp(-6*t)))*np.exp(-t)
number_of_samples = 200
time_domain_bound = 20
transform_domain_bound = 25
```

```python
compz = complex(0, 1)
w = np.linspace(-transform_domain_bound, transform_domain_bound, number_of_samples)
HZ1 = np.zeros((number_of_samples), dtype=complex)
sampling_rate = (transform_domain_bound - (-transform_domain_bound))/number_of_samples
for i in range(number_of_samples):
    real_xt = lambda t: np.real(xt(t) * np.exp(-compz * w[i] * t))
    imag_xt = lambda t: np.imag(xt(t) * np.exp(-compz * w[i] * t))
    h_real = integrate.quad(real_xt, -time_domain_bound, time_domain_bound)
    h_imag = integrate.quad(imag_xt, -time_domain_bound, time_domain_bound) 
    HZ1[i] = (h_real[0] + compz*h_imag[0]) * sampling_rate
```

```python
times = np.linspace(-time_domain_bound, time_domain_bound, number_of_samples)
y = xt(times)
HZ2 = np.fft.fft(y)
```

```python
plt.plot(w, np.real(HZ1), color='cyan', label='Continuous fourier transform')
plt.plot(w, np.real(HZ2), color='yellow', label='fast fourier transform')
plt.legend()
plt.show()
```

```python
def inverse_fourier_transform(hz):
    xn = np.zeros((hz.shape[0]), dtype=complex)
    for i in range(number_of_samples):
        real_hz = lambda w_id: np.real(hz[w_id] * np.exp(compz * w[w_id] * times[i]))
        imag_hz = lambda w_id: np.imag(hz[w_id] * np.exp(compz * w[w_id] * times[i]))
        real_hz_values = real_hz(np.arange(number_of_samples))
        imag_hz_values = imag_hz(np.arange(number_of_samples))
        x_real = integrate.simpson(real_hz_values)
        x_imag = integrate.simpson(imag_hz_values)
        xn[i] = (x_real + compz*x_imag) / (2*np.pi)
    return xn
```

```python
xn1 = inverse_fourier_transform(HZ1) 
xn2 = np.fft.ifft(HZ2)
```

```python
plt.ylim(-1, 3)
plt.plot(times, np.real(xn1), color='magenta', label='written fourier transform')
plt.plot(times, np.real(y) + 0.1, color='cyan', label='original function (+0.1)')
plt.plot(times, np.real(xn2) + 0.2, color='yellow', label='fast fourier transform (+0.2)')
plt.legend()
plt.show()
```
