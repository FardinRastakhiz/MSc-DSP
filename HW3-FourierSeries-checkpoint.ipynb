```python
import numpy as np
import scipy
from scipy import signal, integrate
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
freq1 = 5
freq2 = 10
f = np.gcd(freq1, freq2)
T = 1/f
w0 = 2 * np.pi/T

xt = lambda t: np.sin(2*np.pi*freq1*t) + 2 * np.cos(2*np.pi*freq2*t)

sampling_rate = 200
number_of_samples = 200
```

```python
HZ1 = np.zeros((sampling_rate), dtype=complex)
compz = complex(0, 1)
for i in range(sampling_rate):
    #print(i)
    hz = lambda t: xt(t) * np.exp(-compz * i * w0 * t)
    hzr = lambda t: np.real(hz(t))
    hzi = lambda t: np.imag(hz(t))
    hzr_integral = integrate.quad(hzr, 0, T)
    hzi_integral = integrate.quad(hzi, 0, T)
    hz = hzr_integral[0] + compz * hzi_integral[0]
    HZ1[i] = f*hz*sampling_rate
```

```python
y = xt(np.linspace(0, T, sampling_rate))
HZ2 = np.fft.fft(y)
```

```python
w0_1 = 2*np.pi*freq1
w0_2 = 2*np.pi*freq2

def part1(k):
    denominator = 2*(w0_1**2-k**2*w0**2)
    if denominator==0:
        return 1
    else:
        return -((w0_1 - k*w0)*np.cos(T*w0_1+T*k*w0)+(w0_1+k*w0)*np.cos(T*w0_1-T*k*w0)-2*w0_1)/denominator

    
def part2(k):
    denominator = 2*(w0_1**2-k**2*w0**2)
    if denominator==0:
        return 1
    else:
        return -((w0_1-k*w0)*np.sin(T*w0_1+T*k*w0)+(-w0_1-k*w0)*np.sin(T*w0_1-T*k*w0))/denominator
    
def part3(k):
    denominator = w0_2**2-k**2*w0**2
    if denominator==0:
        return 1
    else:
        return ((w0_2-k*w0)*np.sin(T*w0_2+T*k*w0)+(w0_2+k*w0)*np.sin(T*w0_2-T*k*w0))/denominator
    
def part4(k):
    denominator = w0_2**2-k**2*w0**2
    if denominator==0:
        return 1
    else:
        return -((w0_2-k*w0)*np.cos(T*w0_2+T*k*w0)+(-w0_2-k*w0)*np.cos(T*w0_2-T*k*w0)+2*k*w0)/denominator
#part2 = 
#integrated = lambda k, t: f*(-np.exp(compz*(w0_1-w0*k)*t)/(2*(w0_1-w0*k))-np.exp(-compz*(w0_1+w0*k)*t)/(2*(w0_1+w0*k))-compz*np.exp(compz*(w0_2-w0*k)*t)/(2*(w0_2-w0*k))+compz*np.exp(-compz*(w0_2+w0*k)*t)/(2*(w0_2+w0*k)))
HZ3_lambda = lambda k: (part1(k) + part3(k)) - compz*(part2(k) + part4(k))
```

```python
HZ3 = np.zeros((sampling_rate), dtype=complex)
for i in range(sampling_rate-1):
    #print(f'{i}: {w0_1**2-i**2*w0**2}')
    HZ3[i] = sampling_rate*HZ3_lambda(i)
```

```python
fig, ax = plt.subplots(3, sharex=True)
fig.suptitle('Convolution')
ax[0].plot(np.real(HZ1), color='cyan')
ax[0].set_title("Self writter Fourier Series")
ax[1].plot(np.real(HZ2), color='pink')
ax[1].set_title("Fast Fourier Transform")
ax[2].plot(np.real(HZ3), color='pink')
ax[2].set_title("FManually calculated Fourier Series")
plt.show()
```

```python
def sampling_of_fourier_series(t, hz):
    n_samples = hz.shape[0]
    xn = complex(0, 0)
    for k in range(n_samples):
        xn += hz[k]*np.exp(compz*k*w0*t)
    return xn
#sampling_of_fourier_series(HZ1)
```

```python
t_values = np.linspace(0, T, sampling_rate)
xn1 = sampling_of_fourier_series(t_values, HZ1)
xn2 = sampling_of_fourier_series(t_values, HZ2)
xn3 = sampling_of_fourier_series(t_values, HZ3)


fig, ax = plt.subplots(3, sharex=True)
fig.suptitle('Convolution')
ax[0].plot(xn1, color='cyan')
ax[0].set_title("Self writter Inverse Fourier Series")
ax[1].plot(xn2, color='pink')
ax[1].set_title("Inverse Fast Fourier Transform")
ax[2].plot(xn3, color='magenta')
ax[2].set_title("Manually calculated Fourier Transform")
plt.show()
```
